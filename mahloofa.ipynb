#super market

Introduction to Supermarket Sales Analysis:
 * Supermarket sales analysis is a crucial aspect of retail management, involving the study and interpretation of transaction data to gain insights into customer behavior, product performance, and operational efficiency. By examining various factors such as purchase patterns, customer demographics, and payment methods, businesses can make informed decisions to enhance their marketing strategies, inventory management, and overall customer satisfaction.






The dataset consists of 1000 entries and includes 17 columns. Here are the details:

Columns:
* Invoice ID: Unique identifier for each transaction.
* Branch: Branch where the transaction occurred.
* City: City of the branch.
* Customer type: Type of customer.
* Gender: Gender of the customer.
* Product line: Category of the purchased product.
* Unit price: Price per unit of the product.
* Quantity: Number of units purchased.
* Tax 5%: 5% tax on the total cost .
* Total: Total cost including tax.
* Date: Date of the transaction.
* Time: Time of the transaction.
* Payment: Mode of payment.
* cogs: Cost of goods sold.
* gross margin percentage: Gross margin percentage.
* gross income: Gross income from the transaction.
* Rating: Customer rating of the service.

```python
def label_encode_categorical(df):
    label_encoders = {}
    for col in df.columns:
        if df[col].dtype == 'object':
            label_encoders[col] = LabelEncoder()
            df[col] = label_encoders[col].fit_transform(df[col])
    return df, label_encoders

df, encoders = label_encode_categorical(df)
```
